```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
leads_df = pd.read_csv('Leads.csv')
leads_df.head()
```

```python
leads_df.columns
```

```python
# There are 9240 entries lets check the uniqueness of Prospect ID vs Lead Number
print(leads_df['Prospect ID'].value_counts().sum())
print(leads_df['Lead Number'].value_counts().sum())
```

```python
# So both are unique but Lead Number is an int64, from now considered as a unique entry
# Dropping Prospect ID
leads_df.drop(['Prospect ID'], axis=1, inplace=True)
leads_df.head()
```

```python
# Leads Origin
print(leads_df['Lead Origin'].value_counts())
# All Seems to be in Place
```

```python
# Lead Source Null Value %age
leads_df['Lead Source'].isnull().sum()/leads_df.shape[0]*100
# It is very small but it is a very important metric
```

```python
leads_df['Lead Source'].value_counts()
```

```python
#Looking at Entries with null value in Lead Source
leads_df[leads_df['Lead Source'].isnull()]
```

```python
# Since the conversion Rate is more than 80% we cannot ignore the data
leads_df[leads_df['Lead Source'].isnull()].Converted.mean()*100
```

```python
# Changing NaN Values to Unknown
leads_df['Lead Source'].replace(np.nan,'Unknown',inplace=True)
leads_df['Lead Source'].value_counts()
```

```python
# Google Small and google should be same.
leads_df['Lead Source'].replace('google','Google',inplace=True)
leads_df['Lead Source'].value_counts()
```

```python
# Replacing All Entries below 10 to single group other.
leads_df['Lead Source'].replace(['bing','Click2call','Press_Release','Social Media','Live Chat'
                                 ,'youtubechannel','testone','Pay per Click Ads','welearnblog_Home'
                                 ,'WeLearn','blog','NC_EDM'],'Other',inplace=True)
leads_df['Lead Source'].value_counts()
```
